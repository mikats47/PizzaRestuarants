```
def getPizzaList(city):
  rec = pizza.loc[pizza['city']==city].sort_values(ascending=False, by=['priceRangeMax'])
  return rec[['name', 'priceRangeMax', 'address']].drop_duplicates(subset=['name'], keep=False).head(40)
  
print(getPizzaList('New York'))
```
