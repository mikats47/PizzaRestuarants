```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# set the plots to display in the Jupyter notebook
%matplotlib inline

# change plotting colors per client request
plt.style.use('dark_background')

# Increase default figure and font sizes for easier viewing.
plt.rcParams['figure.figsize'] = (18, 6)
plt.rcParams['font.size'] = 18


pizza = pd.read_csv("./pizza_big.csv")





```

```
pizza.groupby('province')['priceRangeMax'].max().plot(kind='bar')
```

```
pizza.groupby('province')['name'].count().plot(kind='bar')
```

```
pizza.groupby(['name'])['id'].count().sort_values(ascending=False).head(50).plot(kind='bar')
```

```
def getPizzaList(city):
  rec = pizza.loc[pizza['city']==city].sort_values(ascending=True, by=['priceRangeMax'])
  return rec[['name', 'priceRangeMax', 'address']].drop_duplicates(subset=['name'], keep=False).head(40)

city = input('Enter city: ')

print(getPizzaList('New York'))

```
